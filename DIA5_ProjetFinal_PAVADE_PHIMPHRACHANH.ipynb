Ces données ont été préparées pour analyser les facteurs liés à la réadmission ainsi que d'autres résultats concernant les patients diabétiques.

## Où sont collectées les données ?
L'ensemble de données représente 10 années (1999-2008) de soins cliniques dans 130 hôpitaux américains et réseaux de prestation intégrés. Il comprend plus de 50 caractéristiques représentant les résultats des patients et des hôpitaux. Les informations ont été extraites de la base de données pour les rencontres répondant aux critères suivants.
(1) Il s'agit d'une rencontre avec un patient hospitalisé (admission à l'hôpital).
(2) Il s'agit d'une rencontre avec un diabétique, c'est-à-dire une rencontre au cours de laquelle un quelconque type de diabète a été saisi dans le système en tant que diagnostic.
(3) La durée du séjour a été d'au moins 1 jour et d'au plus 14 jours.
(4) Des tests de laboratoire ont été effectués au cours de la visite.
(5) Des médicaments ont été administrés au cours de la visite.
Les données contiennent des attributs tels que le numéro du patient, sa race, son sexe, son âge, le type d'admission, le temps passé à l'hôpital, la spécialité médicale du médecin qui l'a admis, le nombre de tests de laboratoire effectués, le résultat du test HbA1c, le diagnostic, le nombre de médicaments, les médicaments pour le diabète, le nombre de consultations externes, internes et d'urgences dans l'année précédant l'hospitalisation, etc.

Les données Health Facts que nous avons utilisées étaient un extrait représentant 10 ans (1999-2008) de soins cliniques dans 130 hôpitaux et réseaux de prestation intégrés à travers les États-Unis : Midwest (18 hôpitaux), Nord-Est (58), Sud (28) et Ouest (16). La plupart des hôpitaux (78) ont une taille de lit comprise entre 100 et 499, 38 hôpitaux ont une taille de lit inférieure à 100 et 14 hôpitaux ont une taille de lit supérieure à 500.

## Problématique ?

Problem Statement and Objective A hospital readmission is when a patient who is discharged from the hospital, gets re-admitted again within a certain period of time. Hospital readmission rates for certain conditions are now considered an indicator of hospital quality, and also affect the cost of care adversely. For this reason, Centers for Medicare & Medicaid Services established the Hospital Readmissions Reduction Program which aims to improve quality of care for patients and reduce health care spending by applying payment penalties to hospitals that have more than expected readmission rates for certain conditions. Although diabetes is not yet included in the penalty measures, the program is regularly adding new disease conditions to the list, now totaling 6 for FY2018. In 2011, American hospitals spent over $41 billion on diabetic patients who got readmitted within 30 days of discharge. Being able to determine factors that lead to higher readmission in such patients, and correspondingly being able to predict which patients will get readmitted can help hospitals save millions of dollars while improving quality of care. So, with that background in mind, we used a medical claims dataset (description below), to answer these questions:
1. What factors are the strongest predictors of hospital readmission in diabetic patients?
2. How well can we predict hospital readmission in this dataset with limited features?

Comme environ 50% des données appartiennent à la classe "NON", et les autres classes ont moins de données d'individus.
Cela va créer un problème de déséquilibre de classe. Nous allons donc considérer ce problème comme un problème à deux classes.
Nous essaierons seulement de prédire si le patient sera réadmis ou non, nous en séparerons de l'information si le patient a été réadmis de moins de 30 jours ou de plus de 30 jours.
- Création d'une étiquette à 2 classes : Création d'une autre étiquette pour associer <30 et >30 à une classe pour une meilleure analyse et classification.

```python
# Créer une autre colonne et la prendre comme un problème à 2 classes, étiqueter les <30 et >30 comme OUI et les autres "N0" comme NON.

def clean_readmission(value):
    if value == '>30' or value =='<30':
        return 'Yes'
    else:
        return 'No'
    
diabetic_set['readmitted'] =diabetic_set['readmitted'].apply(clean_readmission) 
```

## Diag

Dans les colonnes diag nous avons des valeur qui sont enfaite des str : '?', 'E' et 'V'. Nous avons aussi plusieurs valeurs qui sont enfaite des codes à 3 digit qui représente différentes catégories de maladie.

Il faut donc que remplacer les str par des valeurs numériques puis arrondir les valeur et ensuite placer chaque patient dans une catégorie en fonction de son diagnostique.

```python
diabetic_set = diabetic_set.replace(regex=r'^[E,V].*', value=0)
```

```python
diag_list = ['diag_1', 'diag_2', 'diag_3']
```

```python
#dans chaque colonne diag on remplace les valeurs Nan par -1
for diag in diag_list:
    diabetic_set[diag].fillna(-1, inplace = True)
```

```python
#dans chaque colonne diag on change le type des valeurs de la colonne
for diag in diag_list:
    diabetic_set[diag] = diabetic_set[diag].astype(float)
```

```python
#On créer des catégories pour chaque code
#pour connaitre les catégories on peut s'aider du site suivant : https://www.hindawi.com/journals/bmri/2014/781670/tab2/

def transformCategory(value):
    if value>=390 and value<=459 or value==785:
        return 'Circulatory'
    elif value>=460 and value<=519 or value==786:
        return 'Respiratory'
    elif value>=520 and value<=579 or value==787:
        return 'Digestive'
    elif value==250:
        return 'Diabetes'
    elif value>=800 and value<=999:
        return 'Injury'          
    elif value>=710 and value<=739:
        return 'Musculoskeletal'   
    elif value>=580 and value<=629 or value==788:
        return 'Genitourinary'
    elif value>=140 and value<=239 :
        return 'Neoplasms'
    elif value==-1:
        return 'NaN'
    else :
        return 'Other'
```

```python
#On applique la transformation dans toutes les colonnes diag
for diag in diag_list:
    diabetic_set[diag] = diabetic_set[diag].apply(transformCategory)
```

## Accuracy, Precision and Recall

On se concentre sur l'accuracy et le recall. Pourquoi ?

L'accuracy est important pour savoir si le modèle prédit bien les valeurs.

Quand on va si les patients seront réadmis après leur traitement on va avoir deux situations.

    Situation 1 : Le modèle prédit que le patient sera réadmis (Predicted Positive) mais en fin de compte il n'est pas réadmis (Actual Negative)  => False Positive
    Situation 2 : Le modèle prédit que le patient ne sera pas réadmis (Predicted Negative) mais finalement il est réadmis (Actual Positive)  => False Negative

La situation 1 n'est pas très grave mais la situation 2 est à éviter car cela signifiera que notre modèle n'est pas correct.
On doit se concentrer sur le recall de nos modèles car il y un grand coût associé à la situation 2 (False Negative).
